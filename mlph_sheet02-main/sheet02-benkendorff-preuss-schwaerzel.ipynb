```python
def biweight(x, mu, w):
    """biweight kernel at mean mu, with bandwidth w evaluated at x"""
    #TODO: implement the quartic (biweight) kernel
    mask = np.where(np.abs(x-mu) < w)
    ret = np.zeros(x.shape)
    ret[mask] = 15/(16*w)*(1-((x-mu)[mask]/w)**2)**2
    return ret
```

## 3 Mean-Shift
(a)
The formula for updates to the local mean $x_j^{t+1}=x_j^t+\frac{1}{n}\sum_{i:x_i\in N(x_j)}(x_i-x_j^t)$ is recovered for $\alpha_j^t=\frac{1}{2}$.

This is a sensible learning rate as it is
 - low enough to not skip any maxima
 - high enough to converge in a meaningful time.

```python
# TODO: implement the update to the local mean

from scipy.spatial import distance_matrix

def mean_shift_step(x, xt, r=1):
    """
    A single step of mean shift, moving every point in xt to the local mean of points in x within a radious of r.
    
    Parameters
    ----------
    x : np.ndarray
        Array of points underlying the KDE, shape (d, N1)
    xt : np.ndarray
        Current state of the mean shift algorithm, shape (d, N2)
    n_components : int, optional
        Number of requested components. By default returns all components.
        
    Returns
    -------
    np.ndarray
        the points after the mean-shift step
    """
    # NOTE: For the excercise you only need to implement this for d == 1. 
    #       If you want some extra numpy-practice, implement it for arbitrary dimension
    
    #if x or xt are a vector of size n, transform to 1xn-matrix for scipy to work
    if x.ndim == 1:
        x = np.matrix(x)
    if xt.ndim == 1:
        xt = np.matrix(xt)
    
    assert xt.shape[0] == x.shape[0], f'Shape mismatch: {x.shape[0]}!={xt.shape[0]}'
    
    # TODO: compute a N by N matrix 'dist' of distances, 
    #       such that dists[i, j] is the distance between x[i] and xt[j]
    # this may be considered cheating:
    distMat = distance_matrix(x.transpose(),xt.transpose()) #scipy.spatial.distance_matrix    
    # TODO: threshold the distances with r to get an array of masks for every data point
    mask = np.where(distMat < r)
    # TODO: compute the number of points in x within radius r of each xt
    maskMat = np.zeros(distMat.shape)
    maskMat[mask] = 1
    #print(maskMat.shape, maskMat)
    n_neighbors = np.sum(maskMat, axis=1)
    # TODO: compute the local means by summing over the neighbors of each element in xt 
    #       and dividing by the number of neighbors
    local_means = np.sum(np.where(distMat < r, distMat, 0),axis=1)/n_neighbors
    return local_means
    
```

## 4 K-Means

(a) We assume that an initial guess for the cluster centeres $C\in\mathbb{R}^{p\times K}$ is given, that is, we only formulate the update-steps.

### 1) Updating $m_{kn}$
The first step updates the cluster assignments. For each point $x_i\in\mathbb{R}^p$, we calculate the (euclidean) distance $\|x_i-m_j\|$ for all $j\in\{1,...,K\}$ and then assign $x_i$ to the cluster for which this distance is minimized. I.e. the update step is:

$$m_{ki}=\begin{cases}1\quad,k=\min_l\|x_i-c_l\|\\0\quad,\text{otherwise}\end{cases}$$

In other words: Each point is assigned to the cluster with the cluster-center closest to itself.

### 2) Updating $c_k$
To minimize $E(C,M;K)$, i.e. the sum of (euclidean) distances from all points to their respective cluster centers, it is sufficient to move the "cluster centers" $c_k$ to the actual center of the cluster (which may have changed due to changed cluster assignments, which occured in step 1). Therefore, for alll $k\in\{1,...K\}$, we update:

$$c_k=\frac{1}{\sum_{n=1}^{N}m_{kn}}\sum_{n=1}^{N}m_{kn}x_n,$$
where the denominator $\sum_{n=1}^{N}m_{kn}$ is just the number of datapoints in the $k$-th cluster.

```python
from sklearn.cluster import KMeans

def kmeans_step(x, k, c=None, init='random'):
    """
    K-Means clustering on a p x N data matrix.
    
    Parameters
    ----------
    x : np.ndarray
        Data matrix of shape (p, N).
    k : int
        Number of cluster.
    c : np.ndarray, optional
        Current cluster centers. If None, the initialization as specified by 'init' will be used.
    init: str
        The initialization method to be used if c is None.
        
    Returns
    -------
    float, np.ndarray
        cost after the step, updated cluster centers
    """
    
    #if c is not None: 
    #    assert c.shape[1] == k
        
    # Note: the documentation is your friend: 
    #       https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html
    
    # TODO: set n_cluster, init, n_init and max_iter appropriately
    single_step_kmeans = KMeans(
        n_clusters=k, 
        init=init, 
        max_iter=1,
        n_init=1,
    )
    
    # TODO call the single_step_kmeans
    single_step_kmeans.fit(x)
    
    # TODO get the current energy (you don't have to compute it - it's stored in single_step_kmeans)
    E = single_step_kmeans.inertia_ # this is the member we want according to documentation
    
    # TODO read out cluster centers 
    c = single_step_kmeans.cluster_centers_
    
    return E, c 
    
```

```python
# load the data (you can try both with the full normalized feature and the 2d umap projection)

features = np.load('data/dijet_features_normalized.npy')  # full features
# features = np.load('data/dijet_features_umap.npy')      # umap projection

print(f'{features.shape=}')
```

```python
def kmeans(x, k, max_steps=100, init='random'):
    c = None
    energies = []  # list of energies over the iterations. 
    for i in range(max_steps):
        # Call the kmeans_step implemented above to get the energy and the next cluster centers
        E, c = kmeans_step(x, k=k, c=c, init=init)
        energies.append(E)
        
        # Stop the loop if there was no improvement
        if i>=2 and energies[-1] == energies[-2]:
            break
    energies = np.array(energies)
    return energies  # return array of energies


for k in [3, 5, 10, 20]:
    for init in ['k-means++', 'random']:
        # TODO: for the given k and init, run k-means 20 times or more (using the kmeans function above)
        #       and plot the resulting energy trajectores
        energies = kmeans(features, k, init=init)
        plt.plot(energies, label='Initialization: ' + init)
    plt.title(f'Energy of K-Means for {k} clusters')
    plt.xlabel('Step')
    plt.ylabel('Energy')
    plt.legend(loc='best')
    plt.show()
```

# 5. Bonus: KDE Bandwith and Modes
We found a counterexample with three points $x_1$, $x_2$ and $x_3$ in $\mathbb{R}^2$, for which the KDE has 3 modes at $w=0.1$ and 4 modes at $w=1$. The idea is to position the three datapoints in an equidistant fashion from the origin and each other, so that at $w=0.1$, we have 3 modes (at each position) and at $w=1$, an additional mode is seen at the origin, where the three kernels overlap just so, that $f(x_0)=f(x_1)=f(x_2)=f(x_3)$.

We use a simplified calculation, where we assume that $k(x_i-x_j,1)\approx 0$. Then we have to choose r so that $k(r;w=1)=\frac{1}{3}k(0;w=1)$ (note: the actual value of $w$ does not matter here and is arbitrarily chosen to be 1). We get $r=\sqrt{2\ln(3) w^2}$. A set of points on the unit circle with equal distance between all points is given by $x_1=(0,1)$, $x_2=(\frac{4\sqrt{4}}{10},-\frac{2}{10})$ and $x_3=(-\frac{4\sqrt{4}}{10},-\frac{2}{10})$.

At $w<1$, we have just three modes at $x_1$, $x_2$ and $x_3$. At $w=1$, we have four modes at $x_1$, $x_2$, $x_3$ and the origin.

Below is a visualization of this.

```python
def rbf(x,w, offs=0):
    dist = np.abs(x-offs)
    return 1/(w*np.sqrt(2*np.pi))*np.exp(-np.sum(dist,axis=1)**2/(2*w**2))

xy = np.mgrid[-5:5.1:0.1, -5:5.1:0.1].reshape(2,-1).T

d= np.sqrt(2*np.log(3))

o1 = d*np.array([0,1])
o2 = d*np.array([4*np.sqrt(4)/10,2/10])
o2 = d*np.array([-4*np.sqrt(4)/10,-2/10])
o3 = d*np.array([1,0])

#print(np.sum(np.abs(o1-o2)))
#print(np.sum(np.abs(o1-o3)))
#print(np.sum(np.abs(o3-o2)))
#print(np.sum(np.abs(o1)))
#print(np.sum(np.abs(o2)))
#print(np.sum(np.abs(o3)))

w=1
vals1 = rbf(xy,w, o1)
vals2 = rbf(xy,w,o2)
vals3 = rbf(xy,w,o3)

plt.scatter(xy[:,0],xy[:,1],c=vals1+vals2+vals3)
plt.colorbar()
plt.title('w=1, 4 modes')
plt.show()

w=0.1

vals1 = rbf(xy,w, o1)
vals2 = rbf(xy,w,o2)
vals3 = rbf(xy,w,o3)

plt.scatter(xy[:,0],xy[:,1],c=vals1+vals2+vals3)
plt.colorbar()
plt.title('w=0.1, 3 modes')
plt.show()
```
